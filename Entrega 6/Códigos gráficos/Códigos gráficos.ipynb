```python
df = pd.read_csv("diputados electo.csv")
```

```python
df.columns = df.columns.str.strip()
```

### Gráfico de Parlamentarios Electos (1989-2013)

```python
df_1989_2013 = df[(df['año elección'] >= 1989) & (df['año elección'] <= 2013)]

chart_1989_2013 = alt.Chart(df_1989_2013).mark_bar().encode(
    x=alt.X('año elección:N', title='Año de Elección'),
    y=alt.Y('count():Q', title='Número de Parlamentarios'),
    color=alt.Color('coalición:N', title='Pacto Electoral'),
    tooltip=['año elección', 'coalición', 'count()']
).properties(
    title='Número de Parlamentarios por Pacto Electoral y Año de Elección (1989-2013)'
).interactive()

chart_1989_2013
```

### Gráfico de Parlamentarios Electos (2017-2025)

```python
df_2017_2025 = df[(df['año elección'] >= 2017) & (df['año elección'] <= 2025)]

chart_2017_2025 = alt.Chart(df_2017_2025).mark_bar().encode(
    x=alt.X('año elección:N', title='Año de Elección'),
    y=alt.Y('count():Q', title='Número de Parlamentarios'),
    color=alt.Color('coalición:N', title='Pacto Electoral'),
    tooltip=['año elección', 'coalición', 'count()']
).properties(
    title='Número de Parlamentarios por Pacto Electoral y Año de Elección (2017-2025)'
).interactive()

chart_2017_2025
```

```python
df_grouped = df.copy()

# Consolidate 'Partido de la Gente' entries
df_grouped['coalición'] = df_grouped['coalición'].replace(
    ['Partido de la Gente', 'Partido de la Gente '], 'Partido de la Gente (Consolidado)'
)

# Consolidate 'Independientes' entries
independent_keywords = ['Independiente', 'Independientes']
for keyword in independent_keywords:
    df_grouped.loc[df_grouped['coalición'].str.contains(keyword, case=False, na=False), 'coalición'] = 'Independientes (Consolidado)'


chart_grouped_coalitions = alt.Chart(df_grouped).mark_bar().encode(
    x=alt.X('año elección:N', title='Año de Elección'),
    y=alt.Y('coalición:N', title='Pacto Electoral Consolidado'),
    color=alt.Color('año elección:N', title='Año de Elección'),
    tooltip=['año elección', 'coalición', alt.Tooltip('count():Q', title='Número de Parlamentarios')]
).properties(
    title='P'
).interactive()

chart_grouped_coalitions
```

```python
coalition_counts = df['coalición'].value_counts().reset_index()
coalition_counts.columns = ['coalición', 'count']
top_3_coalitions = coalition_counts.head(3)

chart = alt.Chart(top_3_coalitions).mark_bar().encode(
    x=alt.X('coalición:N', sort='-y', title='Coalición'),
    y=alt.Y('count:Q', title='Número de Candidatos Electos'),
    tooltip=['coalición', 'count']
).properties(
    title='Top 3 Coaliciones con Más Candidatos Electos'
)

chart
```

```python
chart_pie = alt.Chart(top_3_coalitions).mark_arc().encode(
    theta=alt.Theta(field="count", type="quantitative"),
    color=alt.Color(field="coalición", type="nominal", title='Coalición'),
    order=alt.Order('count', sort='descending'),
    tooltip=['coalición', 'count']
).properties(
    title='Top 3 Coaliciones con Más Candidatos Electos (Gráfico de Pastel)'
)

chart_pie
```
