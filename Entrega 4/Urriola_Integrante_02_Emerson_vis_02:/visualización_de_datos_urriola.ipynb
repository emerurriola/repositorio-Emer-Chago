```python
df = pd.read_csv("diputados electos final.csv")
```

```python
df.head(11)
```

```python
df.columns = df.columns.str.strip()
```

```python
alt.Chart(df).mark_bar().encode(
x="año elección:N",
y="coalición:N",
color=alt.Color("año elección:N")
)
```

```python
alt.Chart(df).mark_bar().encode(
    x=alt.X('año elección:N', title='Año de Elección'),
    y=alt.Y('count():Q', title='Número de Parlamentarios'),
    color=alt.Color('coalición:N', title='Pacto Electoral'),
    tooltip=['año elección', 'coalición', 'count()']
).properties(
    title='Número de Parlamentarios por Pacto Electoral y Año de Elección'
).interactive()
```
